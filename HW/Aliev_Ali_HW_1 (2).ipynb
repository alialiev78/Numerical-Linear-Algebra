### Problem 3.2 Implement Distance Metrics (9 pts)

KNN uses a chosen distance metric to find the nearest neighbours. We will use following metrics:

- (3 pts) **Manhattan Distance**
- (3 pts) **Euclidean Distance**
- (3 pts) **Cosine Distance**

Implement each one instead of the provided placeholders.

You must use numpy vector and matrix operations - it's strictly forbidden to use any forms of python loops (for, while, etc.). Implementation of each metric must follow respective placeholder description.

```python
def manhattan_distance(x1, x2):
    '''
    Find the Manhattan distance between two sets of points
        x1 - array of shape (n_samples, n_features)
        x2 - array of shape (m_samples, n_features)

    Returns:
        distance - array of shape (n_samples, m_samples)
    '''
    distance = np.sum(np.abs(x1[:, np.newaxis, :] - x2), axis=2)
    return distance

def euclidean_distance(x1, x2):
    '''
    Find the Eucledian distance between two sets of points
        x1 - array of shape (n_samples, n_features)
        x2 - array of shape (m_samples, n_features)

    Returns:
        distance - array of shape (n_samples, m_samples)
    '''
    diff = x1[:, np.newaxis, :] - x2
    distance = np.sqrt((diff**2).sum(axis=2))

    return distance

def cosine_distance(x1, x2):
    '''
    Find the Cosine distance between two sets of points
        x1 - array of shape (n_samples, n_features)
        x2 - array of shape (m_samples, n_features)

    Returns:
        distance - array of shape (n_samples, m_samples)
    '''
    dot_product = np.dot(x1, x2.T)

    norm_x1 = np.linalg.norm(x1, axis=1, keepdims=True)
    norm_x2 = np.linalg.norm(x2, axis=1, keepdims=True)

    cosine_similarity = dot_product/(norm_x1*norm_x2.T)
    distance = 1 - cosine_similarity

    return distance
```

### Problem 3.3 KNN Classifier (8 pts)

Implement a KNN classifier that can use any of the above distance metrics using provided template. The classifier should have methods for fitting the model to the training data, predicting the class labels for the test data, and finding the K-nearest neighbors.

```python
class NearestNeighborsFinder:
    def __init__(self, n_neighbors=5, metric="euclidean"):
        self.n_neighbors = n_neighbors

        # choose metric function
        if metric == "euclidean":
            self._metric_func = euclidean_distance
        elif metric == "manhattan":
            self._metric_func = manhattan_distance
        elif metric == "cosine":
            self._metric_func = cosine_distance
        else:
            raise ValueError("Metric is not supported", metric)
        self.metric = metric

    def kneighbors(self, X, return_distance=False):
        '''
        Find the K-neighbors of a point.
            X - array of shape (n_queries, n_features)
            return_distance - if True, return distances to the neighbors

        Returns:
            indices - array of shape (n_queries, n_neighbors)
            distances - array of shape (n_queries, n_neighbors)
        '''

        # Calculate distances from each query to all training samples
        distances = self._metric_func(X, self._X)

        # Find the indices of the nearest neighbors
        neighbors_idx = np.argsort(distances, axis=1)[:, :self.n_neighbors]

        # Optionally, get the distances to the nearest neighbors
        if return_distance:
            neighbors_distances = np.take_along_axis(distances, neighbors_idx, axis=1)
            return neighbors_idx, neighbors_distances
        return neighbors_idx


    def fit(self, X, y=None):
        '''
        Fit the model using X as training data
            X - array of shape (n_samples, n_features)
            y - array of shape (n_samples)

        Returns:
            self - fitted estimator
        '''
        self._X = X
        self._y = y
        return self

    def predict(self, X):
        neighbors_idx = self.kneighbors(X)
        predictions = []

        # Perform majority voting among neighbors' classes
        for idx in neighbors_idx:
            neighbor_classes = self._y[idx]
            majority_class = np.bincount(neighbor_classes).argmax()
            predictions.append(majority_class)

        return np.array(predictions)
```

Conclusions: \\

Eucledian distance has the most smooth boundary that's why it's useful for continuous (smooth) data . Also it's the most intuitive distance which can be used in low-dimensions tasks. But features with larger ranges can dominate the distance calculation. \\

Manhattan distance is more like distance with piecewise-linear boundaries.
It's also less sensitive to outliers compared to Euclidean distance and more intuitive in high-dimensional spaces as it considers the absolute differences. That's why it's suitable for high-dimensional data and when the data is sparse (and also for tasks where the chosen path greatly influences, such as in routing problems). \\

Cosine Distance normalizes the vectors, measures the cosine of the angle between two non-zero vectors, focusing on their orientation rather than their magnitude. It's useful for high-dimensional spaces where the direction of the data points is more important than their distance, for example, commonly used in text analysis and recommendation systems where the frequency of terms matters more than their absolute counts.

```python
import time
from tqdm import tqdm

def Hessenberg_Transform(A):
    n = A.shape[0]
    H = np.copy(A)  # Start with a copy of A

    for k in range(n - 2):
        # Create the vector x from the sub-diagonal elements
        x = H[k+1:n, k]  # Elements below the k-th column
        # Create the Householder vector
        e = np.zeros_like(x)
        e[0] = np.linalg.norm(x)  # The norm of x
        u = x - e
        u /= np.linalg.norm(u)  # Normalize u

        # Create the Householder matrix
        H_k = np.eye(n)
        H_k[k+1:n, k+1:n] -= 2 * np.outer(u, u)  # Householder reflection

        # Apply the transformation
        H = H_k @ H  # Update H with the Householder transformation

    return H

# Measure computation time for varying sizes of n
sizes = range(2, 500, 10)
times = []

for n in tqdm(sizes):
    A = np.random.rand(n, n)  # Generate a random n x n matrix
    start_time = time.time()
    Hessenberg_Transform(A)  # Transform to Hessenberg form
    end_time = time.time()
    times.append(end_time - start_time)

# Plotting the results
plt.plot(sizes, times, marker='o')
plt.title('Computation Time of Hessenberg Transform vs Matrix Size')
plt.xlabel('Matrix Size (n x n)')
plt.ylabel('Computation Time (seconds)')
plt.grid()
plt.yscale('log')
plt.xticks(sizes)  # Show all sizes on the x-axis
plt.show()
```

We can see that  the computation time should increases $∼ O(n^3) $, since each Householder transformation involves matrix-matrix multiplications with an $n \times n $ matrix

Let's briefly discuss why the Monte Carlo method is interesting. There are a few reasons for that:

1. It can be used to compute a single element of the inverse or a selected group of elements (hard to do with other methods!). 2. By usual argument if variance is finite Monte Carlo converges as $\frac{1}{\sqrt{N}}$ where $N$ is a number of samples (see, for example, [an article on Monte Carlo integration](https://en.wikipedia.org/wiki/Monte_Carlo_integration)).
3. The method is massively parallel: weights of individual walks for any $ij$ can be computed independently.

**b.** (6 + 6 pts) Explain theoretically, why (and when) one would expect error to drop as $1 \big / \sqrt{N}$ where $N$ is a number of samples. Confirm this for small matrices experimentally. A good way to show this explicitly is to produce a plot alike one [available in wikipedia](https://en.wikipedia.org/wiki/Monte_Carlo_integration#Example) for Monte Carlo integration.

**Solution:**
$\frac{1}{\sqrt N}$ follows from CLT/LLN or from $var(S_n/n) = var(\xi)/n$ (where $S_n = ∑^n_{i=1} \xi$), that's $\sigma(S_n/n) = \sqrt{var S_n/n}$ ~ $\frac{1}{\sqrt n}$

```python
def delta_monte_carlo_inversion(A, i, j, N_walks, eps):
    # Compute the exact inverse
    m = A.shape[0]
    B = np.eye(m) - A
    B_inv = np.linalg.inv(B)

    monte_carlo_estimate = MC_solver(A, i, j, N_walks, eps)
    exact_value = B_inv[i, j]
    return abs(exact_value - monte_carlo_estimate)


N_sizes = list(range(10, 1000, 10))
# print(N_sizes)
delta = []
f_N = []
m = 7
G = np.random.randn(m, m)
Q = np.linalg.qr(G)[0]
A = alpha * Q
eps = 0.1
i = 5
j = 4
for el in tqdm(range(len(N_sizes))):
    delta.append(delta_monte_carlo_inversion(A, i, j, N_sizes[el], eps))
    f_N.append(1/np.sqrt(N_sizes[el]))

# print(delta)
plt.figure(figsize=(10, 6))
plt.plot(N_sizes, delta, label='Delta')
plt.plot(N_sizes, f_N, label='$1/\sqrt(N)$')
plt.xlabel('Size of the sample')
plt.ylabel('Delta')
plt.title('Delta vs. Size of the sample')
plt.grid(True)
plt.yscale('log')
plt.show()
```
